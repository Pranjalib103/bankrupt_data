# handling outliers and imputing them with upper and lower bound

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
def impute(x):
    Q3=df[x].quantile(0.75)
    Q1=df[x].quantile(0.25)
    IQR=Q3-Q1
    lb=Q1-1.5*(IQR)
    ub=Q3+1.5*(IQR)
    idx=np.where(df[x]>ub)
    df[x].iloc[idx]=ub
    idx=np.where(df[x]<lb)
    df[x].iloc[idx]=lb
```

```python
impute("Net Income to Total Assets")
impute("Current Liability to Equity")
impute("Debt ratio %")
impute("ROA(A) before interest and % after tax")
```

# Hypothesis Testing:

```python
from scipy.stats import chi2_contingency, ttest_ind

# Example: T-test for numerical features
significant_features = []
for column in df.columns:
    if df[column].dtype in ['int64', 'float64'] and column != 'Bankrupt?':
        stat, p = ttest_ind(df[column][df['Bankrupt?'] == 1],
                            df[column][df['Bankrupt?'] == 0])
        if p < 0.05:
            significant_features.append(column)
            print(f'Feature: {column}, p-value: {p}')
```

#1.created a empty list
#2.iterated each column and checked for datatype is int or float and column not' bankrupt?'
#3.performed null hypothesis on each column with respect to 'bankrupt?' column
#4. compared p value of each column with 0.05 and which  columns  values with less than p value than 0.05 then appended in empty list
#5. this columns which are in the list are the features that influencing bankrupcy..

```python
print(significant_features) #these are the features which are significantly influencing bankrupcy
```
